```python
import pandas as pd

# 讀檔
df1 = pd.read_csv("328\submission_cv_2_xgb_cost_sensitive_top_15_rf_328.csv")
df2 = pd.read_csv("328\submission_cv_3_xgb_cost_sensitive_top_20_rf_328.csv")

# 合併（用 id 對齊）
df = df1.merge(df2, on="id", suffixes=("_model1", "_model2"))

# 找不同的
diff = df[df["gender_model1"] != df["gender_model2"]]
# pd.set_option('display.max_rows', None)
print(diff.shape)
print(diff)
```

```python
# 讀檔
df1 = pd.read_csv("328/submission_cv_1_xgb_cost_sensitive_top_30_rf_328.csv")
df2 = pd.read_csv("test.csv")

# 合併（用 id 對齊）
df = df1.merge(df2, on="id", suffixes=("_model1", "_model2"))

# 找不同的
diff = df[df["gender_model1"] != df["gender_model2"]]
# pd.set_option('display.max_rows', None)
print(diff.shape)
print(diff)
```

```python
# 比較 target (1, 2) 在兩個結果檔中的數量
count_model1 = df["gender_model1"].value_counts().reindex([1, 2], fill_value=0)
count_model2 = df["gender_model2"].value_counts().reindex([1, 2], fill_value=0)

target_compare = pd.DataFrame({
    "model1_count": count_model1,
    "model2_count": count_model2
})
target_compare.index.name = "target"

print(target_compare)
```
